```python
##Code for target mutation in the known sample assay

#Import relevant python libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

#Read the dataset from two different csv files 
df1=pd.read_csv("labels.csv")
df2=pd.read_csv("curves.csv")

#Merge the two different csv files into one
df=pd.merge(df1, df2, how='inner')
df.to_csv('labels_curves.csv')

#Extract the known samples for an (or many) individual target mutation
known=df[(df['condition']=='ConditionA') & (df['sampleID']>='known0') & (df['targetID']=='m1')].iloc[:,5:].T

#Visualize the known samples for an (or many) individual target mutation
plt.figure(1, figsize=(10,9))
for column in known.columns:
    plt.plot(known[column], color='grey', linestyle='solid')
plt.xticks([0, 25, 50, 75, 100, 125, 150, 175], fontsize=18) 
plt.yticks(visible=True, fontsize=18)
plt.xlabel('time', fontsize=18)
plt.ylabel('signal', fontsize=18)
plt.axvline(70, color='darkgreen', linestyle='dashdot', label='time threshold')
plt.axhline(1e5, color='magenta', linestyle='dotted', label='signal threshold')
plt.legend(fontsize=14)
#plt.savefig('known.png')
```
